```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
import pandas as pd
```

```python
df = pd.read_csv("output.csv")
```

```python
headlines=df['Headline']
headlines
```

```python
sia = SIA()
results = []

for line in headlines:
    pol_score = sia.polarity_scores(line)
    pol_score['headline'] = line
    results.append(pol_score)
```

```python
results
```

```python
neg = []
neu = []
pos = []
compound = []
headline = []
for currDict in results:
    for k, v in currDict.items():

        if k == 'neg':
            neg.append(v)
        if k == 'pos':
            pos.append(v)
        if k == 'neu':
            neu.append(v)
        if k == 'compound':
            compound.append(v)
        if k == 'headline':
            headline.append(v)

```

```python
SIA =pd.DataFrame({ 'Headline': headline,
    'Negative': neg,
     'Positive': pos,
     'Neutral': neu,
     'Compound Score': compound, 

    })
```
